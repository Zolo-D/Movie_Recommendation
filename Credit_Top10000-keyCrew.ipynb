```python
df = pd.read_csv('Credit_Top10000.csv')
df.head()
```

```python
def get_director(data_str):
    if isinstance(data_str, float):
        pass
    else:
        values = []
        data_str = ast.literal_eval(data_str)
        if isinstance(data_str, list):  
            for i in data_str:
                if i['job']=='Director':
                    values.append(i['name'])
            return "|".join(values)
        else:
            return None

df['Director'] = df[['crew']].applymap(get_director)
df.head()
```

```python
def get_screenplay(data_str):
    if isinstance(data_str, float):
        pass
    else:
        values = []
        data_str = ast.literal_eval(data_str)
        if isinstance(data_str, list):  
            for i in data_str:
                if i['job']== 'Screenplay':
                    values.append(i['name'])
                elif i['job']== 'Writer':
                    values.append(i['name'])
            return "|".join(values)
        else:
            return None

df['Screenplay'] = df[['crew']].applymap(get_screenplay)
df.head()
```

```python
def get_cinematographer(data_str):
    if isinstance(data_str, float):
        pass
    else:
        values = []
        data_str = ast.literal_eval(data_str)
        if isinstance(data_str, list):  
            for i in data_str:
                if i['job']=='Director of Photography':
                    values.append(i['name'])
            return "|".join(values)
        else:
            return None

df['Cinematographer'] = df[['crew']].applymap(get_cinematographer)
df.head()
```
