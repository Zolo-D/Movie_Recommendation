```python
df = pd.read_csv('Discover_Top10000.csv')
df.head()
```

```python
IDs = list(df['id'])
```

```python
api_key = '3a99c05a3715265405738606e37342c2'
def get_movie_id(IDs):
    
    df_all = pd.DataFrame()
    
    for i in tqdm(IDs):
        
        url = 'https://api.themoviedb.org/3/movie/'+ format(i) + '?api_key='+ api_key
        r = requests.get(url)
        json_data = r.json()
        df_movie = pd.DataFrame.from_dict(json_data, orient='index').T
        df_all = df_all.append(df_movie, ignore_index=True)
        time.sleep(0.01)
    
    return df_all

if __name__ == '__main__':
    
    df_movie = get_movie_id(IDs)
```
